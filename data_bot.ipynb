```python
import os
import pandas as pd

dataset_path = 'dataset/CrisisLexT26'

def main():
	disasters = os.listdir(dataset_path)

	total_relevant_count = total_irrelevant_count = 0

	final_file = open('disaster_data_relevance.csv', 'w+', encoding='utf-8')
	
	for disaster in disasters:
		
		disaster_folder = dataset_path + '/' + disaster
		files = os.listdir(disaster_folder)
		
		for name in files:
			
			if 'labeled' in name:
				
				tweets = pd.read_csv(disaster_folder + '/' + name)
				tweet_list = tweets.values.tolist()				

				relevant_count = irrelevant_count = 0
				
				for tweet in tweet_list:

					if tweet[4] == 'Related - but not informative':
						final_file.write('"' + tweet[1] + '","' + 'Irrelevant' + '"\n')
						irrelevant_count += 1
						total_irrelevant_count += 1

					elif tweet[4] == 'Related and informative' and ( tweet[3] in ['Affected individuals', 'Infrastructure and utilities']):
						final_file.write('"' + tweet[1] + '","' + 'Relevant' + '"\n')
						relevant_count += 1
						total_relevant_count += 1

				print(disaster, '\nRelevant tweets:', relevant_count, '\nIrrelevant tweets:', irrelevant_count, "\n")

	final_file.close()

	print('Total relevant tweets:', total_relevant_count)
	print('Total irrelevant tweets:', total_irrelevant_count)



if __name__ == '__main__':
	main()
```
